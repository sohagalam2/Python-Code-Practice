##6. Create two random tensors of shape (2, 3) and send them both to the GPU (you'll need access to a GPU for this). Set torch.manual_seed(1234) when creating the tensors (this doesn't have to be the GPU random seed). The output should be something like:

```python
#random seed call
torch.manual_seed(1234)
#availability check
if torch.cuda.is_available():
    device = "cuda" # Use NVIDIA GPU (if available)
else:
    device = "cpu" # Default to CPU if no GPU is available
print(f"Device: \n{device}")

#create two random tensor on GPU
tensor_one = torch.rand(2,3).to(device)
tensor_two = torch.rand(2,3).to(device) #although same call, yet different random number generates
tensor_one, tensor_two
```

##7. Perform a matrix multiplication on the tensors you created in 6

##8. Find the maximum and minimum values of the output of 7.

##10. Make a random tensor with shape (1, 1, 1, 10) and then create a new tensor with all the 1 dimensions removed to be left with a tensor of shape (10). Set the seed to 7 when you create it and print out the first tensor and it's shape as well as the second tensor and it's shape.

```python
#set the seed
torch.manual_seed(7)
#create random tensor
random_tensor = torch.rand(1,1,1,10)
print(f"Random tensor: \n{random_tensor}")
print(f"Random tensor shape: \n{random_tensor.shape}")
#remove the single dimensions
One_dimension_removed = random_tensor.squeeze()
print(f"Result: \n{One_dimension_removed}")
print(f"Result shape: \n{One_dimension_removed.shape}")
```

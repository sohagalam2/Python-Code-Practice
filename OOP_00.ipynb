**Basics of OOP**

*start by defining classes, which act as blueprints, and then create objects from them.*

Four tenants of OOP

* Encapsulation -  bundle data (attributes) and behaviors (methods) within a class to create a cohesive unit.

* Inheritance - the creation of hierarchical relationships between classes, allowing a subclass to inherit attributes and methods from a parent class.

* Anstraction - hiding implementation details and exposing only the essential functionality of an object.

* Polymorphism - treat objects of different types as instances of the same base type, as long as they implement a common interface or behavior.

*In other programming paradigms, objects only represent the data. In OOP, they additionally inform the overall structure of the program.*



```python
"""While the class is the blueprint, an instance is an object
that’s built from a class and contains real data. An instance
of the Dog class is not a blueprint anymore. It’s an actual dog
with a name, like Miles, who’s four years old.

Put another way, a class is like a form or questionnaire.
An instance is like a form that you’ve filled out with information.
"""

class Dog:
  species = "Canis familiaris" #this is class attribute

  def __init__(self, name, age): #initializes each new instance of the class.
    self.name = name
    self.age = age

Miles = Dog("Miles", 12) #To instantiate the Dog class, I need to pass arguments to the parameters
Rex = Dog("Rex", 8)

"""
An object is mutable if you can alter it dynamically. For example,
lists and dictionaries are mutable, but strings and tuples are immutable.

"""

print(Miles.age)
print(Rex.name)
print(Miles.species)

```

```python
class Dog:
  species = "Canis familiaris"

  def __init__(self, name, age):
    self.name = name
    self.age = age

  #Instance method
  def __str__(self):
    return f"{self.name} is {self.age} years old"

  #Another instance method
  def speak(self, sound):
    return f"{self.name} says {sound}"

Rex = Dog("Rex", 8)
print(Rex.__str__())
print(Rex.speak("ghew ghew"))
print(Rex)
```
